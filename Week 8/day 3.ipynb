##  $$ y=2e^{-x} $$ 

```python
y1=y
y2=2*np.e**(-x)  ## y2=2*np.exp(-2*x)
```

```python
plt.plot(x,y1)
plt.plot(x,y2)
```

```python
plt.plot(x,y1,marker="*")
plt.plot(x,y2)
```

```python
plt.plot(x,y1,label="graph of y=2exp(x)")
plt.plot(x,y2,label="graph of y=2exp(-x)")
```

```python
plt.plot(x,y2,label="graph of y=2exp(-x)")
plt.plot(x,y1,label="graph of y=2exp(x)")

plt.legend(loc=0)
```

```python
plt.plot(x,y2,label="graph of $y=2e^{-x}$")
plt.plot(x,y1,label="graph of $y=2e^x$")

plt.legend(loc=0)
plt.grid(True)
```
